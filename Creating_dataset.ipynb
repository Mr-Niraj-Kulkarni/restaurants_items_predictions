```python
#removing  the StoreCode and Ticketcode columns   
df=df.drop(['StoreCode'],axis=1)

#removing the rows with missing values 
df=df.drop(df[df.PartySize == 'na'].index)
```

calculate the number of customers based on TickCode for  :<br>
1. weekday Lunch<br>
2. weekday Dinner<br>
3. weekend Lunch<br>
4. weekend Dinner<br>



```python
#separate the  rows for weekday lunch
tuesday_lunch=df.loc[(df['Shift']=='Lunch') & (df['Day']=='Tuesday')]

#get unique TickCode for 
tuesday_lunch['TicketCode'].unique()
no_of_customer_on_tuesday_lunch=[]

#calculate the number of customer based on Party Size
for tc in tuesday_lunch['TicketCode'].unique():
    a=tuesday_lunch['PartySize'].loc[tuesday_lunch['TicketCode']== tc]
    no_of_customer_on_tuesday_lunch.append(a.sum()/len(a))
    
no_of_customer_on_tuesday_lunch=int(sum(no_of_customer_on_tuesday_lunch))
```

```python
tuesday_dinner=df.loc[(df['Shift']=='Dinner') & (df['Day']=='Tuesday')]
tuesday_dinner['TicketCode'].unique()
no_of_customer_on_tuesday_dinner=[]
for tc in tuesday_dinner['TicketCode'].unique():
    a=tuesday_dinner['PartySize'].loc[tuesday_dinner['TicketCode']== tc]
    no_of_customer_on_tuesday_dinner.append(a.sum()/len(a))
    
no_of_customer_on_tuesday_dinner=int(sum(no_of_customer_on_tuesday_dinner))
```

```python
wednesday_lunch=df.loc[(df['Shift']=='Lunch') & (df['Day']=='Wednesday')]
wednesday_lunch['TicketCode'].unique()
no_of_customer_on_wednesday_lunch=[]
for tc in wednesday_lunch['TicketCode'].unique():
    a=wednesday_lunch['PartySize'].loc[wednesday_lunch['TicketCode']== tc]
    no_of_customer_on_wednesday_lunch.append(a.sum()/len(a))
    
no_of_customer_on_wednesday_lunch=int(sum(no_of_customer_on_wednesday_lunch))
```

```python
wednesday_dinner=df.loc[(df['Shift']=='Dinner') & (df['Day']=='Wednesday')]
wednesday_dinner['TicketCode'].unique()
no_of_customer_on_wednesday_dinner=[]
for tc in wednesday_dinner['TicketCode'].unique():
    a=wednesday_dinner['PartySize'].loc[wednesday_dinner['TicketCode']== tc]
    no_of_customer_on_wednesday_dinner.append(a.sum()/len(a))
    
no_of_customer_on_wednesday_dinner=int(sum(no_of_customer_on_wednesday_dinner))
print(type(no_of_customer_on_wednesday_dinner))
```

# Expanding  and Randomising  Data from 1-JAN-2019 to 30-JUNE-2019


```python

#set the range of dates 
date_range=pd.date_range(start="01/01/2019",end="06/30/2019")
```

```python
#create a empty DataFrame
dataset=pd.DataFrame(columns=('Date','Shift','Day Type','Day','Time','PartySize','MenuCategory','MenuItem','ItemPrice','ItemQty'))

#list of weekday to generate data based on pattern in the problem 
weekday=['Monday','Tuesday','Thursday','Friday']
```

```python

for date in date_range:
    
    #get the day based on date
    day=calendar.day_name[date.weekday()]
    
    if day in weekday:
        
        #calculate the random no. of customers using assumptions above
        no_of_customer_weekday_lunch=np.random.randint(no_of_customer_on_tuesday_lunch-10,no_of_customer_on_tuesday_lunch+5)
        
        #add rows for weekday lunch 
        add_rows('Lunch',date,day,'Weekday',no_of_customer_weekday_lunch,dataset)
        
        #calculate the random no. of customers using assumptions above
        no_of_customer_weekday_dinner=np.random.randint(no_of_customer_on_tuesday_dinner,no_of_customer_on_tuesday_dinner+15)
        
        #add rows for weekday dinner 
        add_rows('Dinner',date,day,'Weekday',no_of_customer_weekday_dinner,dataset)
        
    else:
        
        #calculate the random no. of customers using assumptions above
        no_of_customer_weekend_lunch=np.random.randint(no_of_customer_on_wednesday_lunch-5,no_of_customer_on_wednesday_lunch+15)
        
        #add rows for weekend lunch 
        add_rows('Lunch',date,day,'Weekend',no_of_customer_weekend_lunch,dataset)
        
        #calculate the random no. of customers using assumptions above
        no_of_customer_weekend_dinner=np.random.randint(no_of_customer_on_wednesday_dinner,no_of_customer_on_wednesday_dinner+30)
        
        #add rows for weekday dinner 
        add_rows('Dinner',date,day,'Weekend',no_of_customer_weekend_dinner,dataset)
```

```python

def add_rows(shift,date,day,day_type,no_of_customer,dataset):
'''
function to add the random data generated to the data frame
'''        
    #generate a random  partySize add to dataset within no_of_customer at particular shift and week of the day  
    while no_of_customer>0:
        
        #get random time 
        time=get_random_time(shift)
        
        party_size=np.random.randint(2,6)
        
        if party_size == 2:
            
            no_of_items=np.random.randint(2,5)
            
        elif party_size == 3:
            
            no_of_items=np.random.randint(4,7)
            
        elif party_size == 4:
            
            no_of_items=np.random.randint(4,8)
         
        else:
            no_of_items=np.random.randint(5,8)
            
        
        #get random menu Category ,menu Item and Item Qunantity from given data  
        for i in range(no_of_items):
            
            temp=df.loc[df['PartySize']==party_size]
            
            temp1=temp.iloc[random.choice(range(len(temp)))]
            menu_category=temp1['MenuCateogry']
            
            menu_item=temp1['MenuItem']
            
            item_price=temp1['ItemPrice']
            item_quantity=temp1['ItemQty']
            
            #add to the the dataset
            dataset.loc[len(dataset)+1] = [date,shift,day_type,day,time,party_size,menu_category,menu_item,item_price,item_quantity]
        
        
        no_of_customer-=party_size
```

```python
def get_random_time(shift):
'''
function to get random time based on range of Lunch and Dinner timmings
'''    
    
    if shift=='Lunch':
        
        start_string = '11:30 AM'
        end_string='02:00 PM'
        
    elif shift=='Dinner':
        
        start_string = '06:30 PM'
        end_string='10:00 PM'

    stime= pd.to_datetime(start_string)
    etime= pd.to_datetime(end_string)
    
    #generate random time 
    random_time=stime+((etime-stime)*random.random())
    
    return random_time
```
